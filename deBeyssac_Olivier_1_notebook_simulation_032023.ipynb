# Introduction.

Olist, compagnie Brésilenne, se présente comme un eco-système à destination du segment "SMB" et propose des services orientés recherche de capital, prospection de nouveaux clients et organisation des flux logistiques. Ces services s'appuient sur de la donnée régulièrement mise à jour par les équipes d'Olist (https://www.crunchbase.com/organization/olist).

La mission consiste à produire aux équipes d'e-commerce une segmentation des clients qu’elles pourront utiliser au quotidien pour leurs campagnes de communication.

Il s'agit de fournir à l’équipe marketing une description actionable de la segmentation et de sa logique sous-jacente pour une utilisation optimale, ainsi qu’une proposition de contrat de maintenance basée sur une analyse de la stabilité des segments au cours du temps.

Il faut en particulier aider les équipes d’Olist à comprendre les différents types d'utilisateurs. Ces types d'utilisateurs feront l'objet de catégories qui seront utilisées par l’équipe Marketing.

```python
def display_sales(df1, df2):
    l_ =[df1[0:20], df2[0:20]]
    
    fig, axes = plt.subplots(1,2, figsize = (7,4))
    plt.tight_layout(pad = 3.0)
    
    for i, axes in enumerate(axes.flatten()):
        data = l_[i]
        
        sns.barplot(x = l_[i].index,
                   y = l_[i].iloc[:,1], ax = axes)
        axes.set_xticklabels(l_[i].index, fontsize = 6, rotation = 45)
        
    fig.suptitle('Sales & Orders, first 20 customers')
    plt.show()
    return
```

## Cumulative sales, orders.

```python
d = {'order_item_id': lambda item_id: item_id.count(),
     'payment_value': lambda pay_value: pay_value.sum()}


df_agg = df_data.groupby(['order_purchase_timestamp'], as_index = False).agg(d)

df_agg['cum_sales'] = df_agg['payment_value'].cumsum(axis=0)
#df_agg['cum_sales'] = np.log(df_agg['cum_sales'])

df_agg['cum_orders'] = df_agg['order_item_id'].cumsum(axis=0)
#df_agg['cum_orders'] = np.log(df_agg['cum_orders'])

df_cum_sales = df_agg[['order_purchase_timestamp', 'cum_sales']]
df_cum_orders = df_agg[['order_purchase_timestamp', 'cum_orders']]
```

### Vérification, remplacement valeurs manquantes variables de dates.

```python
# Vérification, remplacement valeurs manquantes variables de dates.

def missing_val(df):
    col_name_1 = 'order_delivered_customer_date'
    col_name_2 = 'order_estimated_delivery_date'
    
    
    if df[col_name_1].isnull().any() == True:
        df[col_name_1].isnull().mean()*100
        df[col_name_1].fillna(df[col_name_1].median(), inplace = True)
        
        
        
    if df[col_name_2].isnull().any() == True:
        df[col_name_2].isnull().mean()*100
        df[col_name_2].fillna(df[col_name_2].median(), inplace = True)
        
        

    return print(df_data['order_delivered_customer_date'].isnull().any()), print(df_data['order_estimated_delivery_date'].isnull().any())
```

### Calcul dates d'achat la plus lointaine et la plus récente

```python
# calcul dates d'achat la plus lointaine et la plus récente et liste des colonnes de df.

def min_max_dates(df):
    min_date = df['order_purchase_timestamp'].min()
    max_date = df['order_purchase_timestamp'].max()
    l_cols = list(df_data.columns)
    
    
    return l_cols, min_date, max_date
```

```python
def rfmd(df):

    d = {'order_purchase_timestamp': lambda duration: (max_date - duration.max()).days,
         'order_item_id': lambda item_id: item_id.count(),
         'payment_value': lambda pay_value: pay_value.sum(),
         'order_delivered_customer_date': lambda deliv_cust_date: deliv_cust_date.mean(),
         'order_estimated_delivery_date': lambda estim_cust_date: estim_cust_date.mean()}
     

    rfm = df.groupby(['customer_id'], as_index = False).agg(d)
    
    rfm['extra_days_vs_estim'] = rfm['order_delivered_customer_date'] - rfm['order_estimated_delivery_date']
    rfm['extra_days_vs_estim'] = rfm['extra_days_vs_estim'].astype('timedelta64[D]')

    rfmd = rfm.drop(['order_delivered_customer_date', 'order_estimated_delivery_date'], axis = 1)
    
    return rfmd
```

```python
def rename_cols(df):
    l_old_names = ['order_purchase_timestamp',
                   'order_item_id','payment_value',
                   'extra_days_vs_estim']
    
    l_new_names = ['recency', 'frequency',
                   'monetary', 'extra_days_vs_estim']
    
    
    for name in l_old_names:
        
        i = l_old_names.index(name)
        
        df.rename(columns = {name:l_new_names[i]}, inplace = True)
        
    return df
```

```python
# Visualisation des clusters sous forme de nuages de points et taille 

range_n_clusters = [3, 4, 5, 6, 7, 8]

for n_clusters in range_n_clusters:
    # Création des subplots, 1 ligne, 2 colonnes
    fig, (ax1, ax2) = plt.subplots(1, 2)
    fig.set_size_inches(18, 7)
    
    ax1.set_xlim([-1, 1])
    ax1.set_ylim([0, len(df_projected) + (n_clusters + 1) * 10])
    
    
    # Modèle de clustering
    clusterer = KMeans(n_clusters=n_clusters, n_init="auto", random_state=0)
    
    # Obtention des labels
    cluster_labels = clusterer.fit_predict(X_projected[:,0:3])
    
    # Obtention des coordonnées des centres
    centers = clusterer.cluster_centers_
        
    # Calcul du score du modèle
    silhouette_avg = silhouette_score(X_projected[:,0:3], cluster_labels)
    print(
        "Pour n_clusters =",
        n_clusters,
        "la moyenne du silhouette_score est de :",
        silhouette_avg)
    
    
    sample_silhouette_values = silhouette_samples(X_projected[:,0:3], cluster_labels)
    
    y_lower = 10
    
    
    for i in range(n_clusters):
        
        # Agréger les valeurs de score par cluster + tri valeurs        
        ith_cluster_silhouette_values = sample_silhouette_values[cluster_labels == i]

        ith_cluster_silhouette_values.sort()

        size_cluster_i = ith_cluster_silhouette_values.shape[0]
        y_upper = y_lower + size_cluster_i

        color = cm.nipy_spectral(float(i) / n_clusters)
        ax1.fill_betweenx(
            np.arange(y_lower, y_upper),
            0,
            ith_cluster_silhouette_values,
            facecolor=color,
            edgecolor=color,
            alpha=0.7)

        # Etiqueter les clusters
        ax1.text(-0.05, y_lower + 0.5 * size_cluster_i, str(i))

        # Calculer la nouvelle ordonnée pour le graphe suivant
        y_lower = y_upper + 10  # 10 for the 0 samples

    ax1.set_title("The silhouette plot for the various clusters.")
    ax1.set_xlabel("The silhouette coefficient values")
    ax1.set_ylabel("Cluster label")

    # Visualisation du score selon ligne verticale pour chaque cluster
    ax1.axvline(x=silhouette_avg, color="red", linestyle="--")

    ax1.set_yticks([])  # Enlever les étiquettes sur l'axe des ordonnées (ticks)
    ax1.set_xticks([-0.1, 0, 0.2, 0.4, 0.6, 0.8, 1])

    # Construire le graphe des nuages de points
    colors = cm.nipy_spectral(cluster_labels.astype(float) / n_clusters)
    
    ax2.scatter(x = df_projected['pc0'].values,
                y = df_projected['pc1'].values,
                marker=".",
                s=30,
                lw=0,
                alpha=0.7,
                c=colors,
                edgecolor="k")


    # Dessin des centres de clusters    
    ax2.scatter(x = df_p_centers['pc0'].values,
                y = df_p_centers['pc1'].values,
                marker="o",
                c="white", alpha=1,
                s=200, edgecolor="k")

    for i, c in enumerate(centers[:,0:2]):
        #print(i, c)
        ax2.scatter(c[0], c[1],
                    marker="o",
                    c="white", alpha=1,
                    s=200, edgecolor="k")
        
        ax2.scatter(c[0], c[1],
                    marker="$%d$" % i,
                    alpha=1,
                    s=50, edgecolor="k")

    ax2.set_title("Visualization des clusters.")
    ax2.set_xlabel("Feature space for the 1st feature")
    ax2.set_ylabel("Feature space for the 2nd feature")

    plt.suptitle(
        "Silhouette analysis for KMeans clustering on sample data with n_clusters = %d"
        % n_clusters,
        fontsize=14,
        fontweight="bold",
    )

plt.show()
```

# Cluster maintenance.

Le point ici est d'appréhender à quelle vitesse les clusters évoluent à cause du changement de comportement des clients. Cette modification des clusters devra entrainer une adaptation des tactiques de marketing et commerciales. La méthode consiste à:
- Partir du df_data initial.
- Construire une fonction pour fabriquer une progression dans le temps. Incrément 3 mois par exemple.
- Construire une fonction qui sélectionne les data en fonction de la date incrémentée.
- Construire une fonction qui fabrique les KPI RFM et D.
- Construire une fonction qui centre et fait une mise à l'échelle.
- Construire une fonction qui réalise le clustering des clients en fonction des périodes.
- Afficher les scores ARI.

```python
print(min_date)
print(max_date)

def incremental_dates():
    start_output = []
    end_output = []
    
    n_start_output = []
    n_end_output = []
    
    
    increment = 1
    diff_dates = relativedelta(max_date, min_date)
    diff_months = diff_dates.months + diff_dates.years*12
    
    nb = int(diff_months/increment)
    
    start_date= min_date + relativedelta(months=6)
    
# Calcul dates correspondant aux périodes de référence    
    for i in range(nb):
        ending_date = start_date + relativedelta(months = increment)        
    
        if ending_date <= max_date:
            start_output.append(start_date)
            end_output.append(ending_date)
                        
        start_date = ending_date + relativedelta(days = 1)
                
    outputs = zip(start_output, end_output)
        
    
    return outputs
    
    
    
    
```

```python
def transform(df):
    numerical_features = df.select_dtypes('float64')
    
    num_pipe = make_pipeline(StandardScaler(),SimpleImputer(strategy='mean'))
    num_columns = list(numerical_features.columns)
    
    
    preprocessor = make_column_transformer((num_pipe, num_columns))
    df_scaled = preprocessor.fit_transform(df)
    print("len_df_scaled: {}".format(len(df_scaled)))
    
    return df_scaled
```

```python
output = []

l_labels_true  =[]
l_labels_pred = []

l_ari_values = []


def global_function(df):
    
    start_date = outputs[0][0]
    end_date = outputs[0][1]
    
    mask1 = (df['order_purchase_timestamp'] >= start_date) & (df['order_purchase_timestamp'] < end_date)
#     print("ref {}, {}".format(start_date, end_date))
        
    df1 = df.loc[mask1,:]
    a1 = rfmd(df1)
    b1 = rename_cols(a1)
    c1 = b1.drop('customer_id', axis = 1)
        
    c1_scaled = transform(c1)
    
    
    # Estimator KMeans
    n_init = 'auto'
    algorithm = 'lloyd'
    random_state = 0
    
    kmeans = KMeans(n_init = n_init, algorithm = algorithm, random_state = random_state)
        
    # recherche optimisation nb clusters
    param_grid = {'n_clusters': [i for i in range(3,10)]}
    gscv = GridSearchCV(kmeans, param_grid = param_grid, refit = True).fit(c1_scaled)       
    n_clusters_1 = gscv.best_params_['n_clusters']

    print("n_clusters_1: {}".format(n_clusters_1))
    
    kmeans_1 = KMeans(n_clusters=n_clusters_1,
                      n_init = n_init,
                      algorithm = algorithm,
                      random_state = random_state)
    
    kmeans_1.fit(c1_scaled)
        

        
    for i in range(1, len(outputs)):
        n_start_date = outputs[i][0]
        n_end_date = outputs[i][1]
#         print("pred {},{}, {}".format(i,n_start_date, n_end_date))
            
        mask2 = (df['order_purchase_timestamp'] >= n_start_date) & (df['order_purchase_timestamp'] < n_end_date)

        df2 = df.loc[mask2,:]
        
        if len(df2) != 0:
            a2 = rfmd(df2)
            b2 = rename_cols(a2)
            c2 = b2.drop('customer_id', axis = 1)

            c2_scaled = transform(c2)
        

        
            # Clustering sur périodes t+1 en faisant un ".predict()".
            labels_pred = kmeans_1.predict(c2_scaled)
#             print("labels_pred {}".format(labels_pred))


            # Clustering sur périodes t+1 en faisant un ".fit()".
#             kmeans_2 = KMeans(n_clusters=n_clusters_1,
#                               n_init = n_init,
#                               algorithm = algorithm,
#                               random_state = random_state)
        
            kmeans_1.fit(c2_scaled)
            
            labels_true = kmeans_1.fit(c2_scaled).labels_
#             print("n_clusters_2: {}".format(n_clusters_2))
#             print("labels_pred {}".format(labels_true))
            
            
            # ARI computing.
            ARI = adjusted_rand_score(labels_true, labels_pred)
            
        if len(df2) == 0:
            ARI = 0.5
            
        l_ari_values.append(ARI)
            
    return l_ari_values
```

```python
def display_ARI(array):

    l_index = []
    data = array
    
    for i in range(len(data)):
        index = 'période/dataset: ' + str(i)
        l_index.append(index)
        

        
    df = pd.DataFrame(data=data, index = l_index,columns=['ARI_values'])
    df['MA'] = df['ARI_values'].rolling(6).mean()
    print(df)
    
    figure, axes = plt.subplots(figsize=(6,6))
    sns.lineplot(x=df.index, y='ARI_values', data=df, marker="o", ax=axes)
    sns.lineplot(x=df.index, y=1, markers=True, marker="o", ax=axes)
    sns.lineplot(x=df.index, y=0.5, markers=True, marker="o", ax=axes)
    sns.lineplot(x=df.index, y='MA', data=df, markers=True, marker="o", linestyle='dashed',ax=axes)
    
    plt.title('Score_ARI')
    plt.xticks(fontsize=7, rotation=45)
    plt.yticks(fontsize=7)
    plt.show()
    
    return
```
